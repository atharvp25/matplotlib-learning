```python
import matplotlib.pyplot as plt

x=["Jan","Feb","Mar","Apr","May","Jun"]
y=[1200,1000,1500,2000,1600,2100]

plt.plot(x,y,color="red",linestyle="--",linewidth=2,marker="o",label="Sales Graph")
plt.xlabel("Month")
plt.ylabel("Sales per Month")
plt.legend()
plt.ylim(500,2500)
plt.grid(color="grey",linestyle=":")
plt.show()
```
