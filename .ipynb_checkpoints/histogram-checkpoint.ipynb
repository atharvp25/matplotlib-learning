```python
import matplotlib.pyplot as plt

ages = [22, 25, 27, 25, 30, 32, 34, 35, 36, 36, 
        37, 38, 40, 41, 42, 43, 45, 45, 46, 47, 
        48, 50, 52, 54, 55, 57, 58, 60, 62, 65]

plt.hist(ages,bins=5,color="yellow",edgecolor="black")
plt.xlabel("Ages")
plt.ylabel("No. of People")
#plt.xlim(22,65)
plt.title("AGES and NO. People ")
plt.show()
```
