```python
import matplotlib.pyplot as plt

fig,ax=plt.subplots(2,2)

x=[1,2,3,4,5]
y=[25,10,15,20,25]


ax[0][0].plot(x, y)
ax[0][0].set_title("Line Plot")

ax[0][1].bar(x, y)
ax[0][1].set_title("Bar Graph")

ax[1][0].scatter(x,y)
ax[1][0].set_title("Scatter plot")

fig.suptitle("Comparision")

plt.tight_layout()
```
