```python
#Working With Single Data Set

import matplotlib.pyplot as plt

# Hours Studied
hours = [1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0]

# Test Scores
scores = [52, 55, 60, 62, 65, 70, 72, 78, 82, 88]

plt.scatter(hours,scores,color="orange",marker="o",label="Study and Score")
plt.xlabel("Hours")
plt.ylabel("Score")
plt.grid()
plt.legend()

plt.show()
```

```python
#Working with 2 data set

import matplotlib.pyplot as plt

hours_A = [1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0]
scores_A = [52, 55, 60, 62, 65, 70, 72, 78, 82, 88]

hours_B = [1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0]
scores_B = [50, 55, 60, 65, 70, 75, 82, 88, 90, 98]



plt.scatter(hours_A,scores_A,color="lightblue",marker="o",label="Class A")
plt.scatter(hours_B,scores_B,color="orange",marker="o",label="Class B")

plt.xlabel("Hours")
plt.ylabel("Scores")

plt.grid()
plt.legend()

plt.show()
```
