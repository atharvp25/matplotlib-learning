```python
import matplotlib.pyplot as plt

product = ["A","B","C","D"]
sales = [1000,500,1500,2000]

#For horizontal view 
#plt.barh(product,sales,color="skyblue",label="Product Sales")


plt.bar(product,sales,color="skyblue",label="Product Sales")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.legend()
plt.grid()
plt.title("Product Sales")

plt.show()
```
